```python
# Plotting Libraries
import matplotlib.pyplot as plt
import matplotlib as mpl

# ML and data libraries
from sklearn.mixture import GaussianMixture
import numpy as np
import scipy
import pandas as pd 

# System libraries
import pickle 
```

```python
from matplotlib.patches import Ellipse
def plot_cov_ellipse(cov, pos, nstd=2, ax=None, **kwargs):
    """
    Plots an `nstd` sigma error ellipse based on the specified covariance
    matrix (`cov`). Additional keyword arguments are passed on to the 
    ellipse patch artist.

    Parameters
    ----------
        cov : The 2x2 covariance matrix to base the ellipse on
        pos : The location of the center of the ellipse. Expects a 2-element
            sequence of [x0, y0].
        nstd : The radius of the ellipse in numbers of standard deviations.
            Defaults to 2 standard deviations.
        ax : The axis that the ellipse will be plotted on. Defaults to the 
            current axis.
        Additional keyword arguments are pass on to the ellipse patch.

    Returns
    -------
        A matplotlib ellipse artist
    """
    def eigsorted(cov):
        vals, vecs = np.linalg.eigh(cov)
        order = vals.argsort()[::-1]
        return vals[order], vecs[:,order]

    if ax is None:
        ax = plt.gca()

    vals, vecs = eigsorted(cov)
    theta = np.degrees(np.arctan2(*vecs[:,0][::-1]))

    # Width and height are "full" widths, not radius
    width, height = 2 * nstd * np.sqrt(vals)
    ellip = Ellipse(xy=pos, width=width, height=height, angle=theta, **kwargs)

    ax.add_artist(ellip)
    return ellip
```

```python
file = "../../data/SF.pickle"
SF = pickle.load(open(file, "rb"))
```

```python
mag = SF[0]['magdiff']
time = SF[0]['timediff']
d = len(mag)
a = np.hstack((mag.reshape((d,1)), time.reshape((d,1))))
```

```python

N = len(SF)
K = 20 # number of clusters to estimate

GMM = GaussianMixture(n_components=K, covariance_type='diag')
```

```python
#N = 10 # temperarily set this to small for testing


X_mu = np.zeros((N, K, 2))
X_sigma = np.zeros((N, K, 2))

for i in range(N):
    mag = SF[i]['magdiff']
    time = SF[i]['timediff']
    d = len(mag)
    sf = np.hstack((time.reshape((d,1)), mag.reshape((d,1))))
    GMM = GMM.fit(sf)
    
    X_mu[i,:] = GMM.means_
    X_sigma[i, :, :] = GMM.covariances_
    
```

```python
# split the data
labels = np.zeros(N)
for i in range(N):
    labels[i] = SF[i]['class'] == 'CV'

msk = np.random.rand(N) < 0.8

def F(A):
    (s1,_) = A.shape
    return np.reshape(A, (s1, 2*K))

print(X_mu.shape)
print(F(X_mu).shape)
print(X_sigma.shape)

training_data = np.concatenate(F(X_mu[msk]), F(X_sigma[msk,:]), axis=1)
print(training_data.shape, X_mu.shape)
training_labels = labels[msk]

testing_data = X_mu[~msk,:]
testing_labels = labels[~msk]
```

```python

from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(10), random_state=1)
clf.fit(training_data, training_labels)
pred = clf.predict(testing_data)

accur = np.count_nonzero(pred == np.array(testing_labels)) / len(testing_labels)
```
